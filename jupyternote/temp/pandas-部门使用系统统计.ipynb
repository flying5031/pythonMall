```python
data = pd.read_excel(r'D:\python\pynote\pandas\doc\模拟.xlsx',index_col='支队')
```

```python
def f(x):
    li = ['八支队','九支队']
    for s in li:
        if x.startswith(s):
            return s
    return x
```

```python
df1=data.groupby(f).sum()
```

```python
df1
```

```python
df1.loc['总和']=df1.sum()
```
