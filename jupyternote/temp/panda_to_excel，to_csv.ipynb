# DataFrame输出到excel，to_excel，to_csv方法

`DataFrame.to_excel(excel_writer, sheet_name='Sheet1', na_rep='', float_format=None, columns=None, header=True, index=True, index_label=None, startrow=0, startcol=0, engine=None, merge_cells=True, encoding=_NoDefault.no_default, inf_rep='inf', verbose=_NoDefault.no_default, freeze_panes=None, storage_options=None)`  
**可指定sheet页，可用WriteExcel对象写入多个sheet，可指定是否写入header和index，指定index的名字，指定写几行几列，指定空值替换为什么符号**

1. na_rep:str, default ‘’  
    :Missing data representation.空值默认写空，也可替换为其他  
2. float_format:str, optional  
Format string for floating point numbers. For example float_format="%.2f" will format 0.1234 to 0.12.  
3. indexbool, default True  
Write row names (index). 是否写入索引，默认是  

`DataFrame.to_csv(path_or_buf=None, sep=',', na_rep='', float_format=None, columns=None, header=True, index=True, index_label=None, mode='w', encoding=None, compression='infer', quoting=None, quotechar='"', lineterminator=None, chunksize=None, date_format=None, doublequote=True, escapechar=None, decimal='.', errors='strict', storage_options=None)`

```python
with pd.ExcelWriter(r'D:\python\pynote\pandas\doc\w2_c.xlsx') as writer:
    ex.to_excel(writer,'sheet11')
    df2.to_excel(writer,'sheet22')
```

### 用ExcelWriter写入到两个sheet页 
mode='a' 追加模式，index 是否写入索引，默认为是

```python
with pd.ExcelWriter(r'D:\python\pynote\pandas\doc\w3_c.xlsx' ,engine='openpyxl') as writer:
    ex.to_excel(writer,'sheet11',index=False)
    df2.to_excel(writer,'sheet22',index=False)
```

```python
with pd.ExcelWriter(r'D:\python\pynote\pandas\doc\w4_c.xlsx' ,engine='openpyxl') as writer:
    ex.to_excel(writer,'sheet11',index_label='AB')
    df2.to_excel(writer,'sheet22')
```
